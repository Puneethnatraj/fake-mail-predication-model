```python
nltk.download('stopwords')
```

```python
def process_text(message):
  nopunc=[char for char in message if char not in string.punctuation]
  nopunc=''.join(nopunc)
  clean_words=[word for word in nopunc.split() if word.lower() not in stopwords.words('english')]
  return clean_words
```

```python
df1['message'].head().apply(process_text)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
message_bow=CountVectorizer(analyzer=process_text).fit_transform(df1['message'])
```

```python
message_bow
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(message_bow,df1['file'],test_size=0.20 , random_state=0)
```

```python
message_bow.shape
```

```python
#naive ayes classifier
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB().fit(X_train,Y_train)
```

```python
print(classifier.predict(X_train))
print(Y_train.values)
```

```python
#train_data
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
pred = classifier.predict(X_train)
print(classification_report(Y_train,pred))
print()
print('confusion Matrix: \n', confusion_matrix(Y_train,pred))
print()
print('Acuracy:',accuracy_score(Y_train,pred))
```

```python
print(classifier.predict(X_test))
print(Y_test.values)
```

```python
#test_data
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
pred = classifier.predict(X_test)
print(classification_report(Y_test,pred))
print()
print('confusion Matrix: \n', confusion_matrix(Y_test,pred))
print()
print('Acuracy:',accuracy_score(Y_test,pred))
```
